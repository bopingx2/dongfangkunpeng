#### 2、双变量分析
首先，我们先进行相关性分析。相关性分析主要用于探索两个连续型变量之间的关系。Pearson相关系数是常用的统计量，用于度量两个数值型变量之间的线性关系强度和方向。

```python
# 相关性分析
correlation_matrix = data[['是否逾期','工龄', '年收入log', '征信查询次数', '总贷款额度log', '贷款收入比']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.4)
plt.title('相关性分析-热力图')
plt.show()
```

* **数据解读：**  
1. 总贷款额度log与贷款收入比：相关系数为0.7，表示这两个变量间存在较强的正线性关系。即随着贷款总额的增加，贷款与收入的比值也会增加。  
2. 总贷款额度log与年收入log：相关系数为0.48，表示两者之间有中等程度的正线性关系。这意味着年收入较高的人群可能会有较高的贷款总额。  
3. 是否逾期与其他变量的相关系数都很低，与“工龄‘相关系数为-0.071，显示了一个轻微的负线性关系。这可能意味着工龄较长的个体有较低的逾期风险，但这个相关性较弱。 
* **处理方案：**  
1. 多重共线性：由于“总贷款额度log”与“贷款收入比”的相关系数达到0.7，需要注意多重共线性的问题。在回归模型中，高度相关的自变量可能导致模型的不稳定。可能需要考虑删除其中一个变量或使用降维技术。
2. 非线性关系：尽管“是否逾期”与其他变量的相关系数较低，但这并不意味着它们之间没有其他形式的关系。可能存在非线性关系或其他高阶交互作用。可以考虑使用决策树、随机森林或增强学习等模型捕捉这些非线性关系。
3. 特征工程：考虑构建或提取新的特征，如变量的二次项、交互项等，以提高模型的预测能力。


下面我们进行**分组统计分析** 。   
分组统计分析是为了探索基于某个分类变量（如“职业类别”）时，数值型变量（如“逾期率”）的变化趋势和特性。通过此分析，我们希望确定特定的类别（例如某个职业）是否会有更高或更低的逾期率。

```python
# 1. 计算逾期率

# 根据'职业类别'分组并计算'是否逾期'的均值，这样得到的就是每个职业的逾期率
grouped_data = data[['是否逾期', '职业类别']].groupby('职业类别').mean()

# 将逾期率转为百分比格式
grouped_data['逾期率'] = grouped_data['是否逾期'] * 100

# 重置索引，以便后续使用
ocu_plt = grouped_data.reset_index()

# 将结果按逾期率降序排列，以便后续绘图
ocu_plt.sort_values('逾期率', inplace=True, ascending=False)

# 打印出每个职业类别的逾期率
print(ocu_plt[['职业类别', '逾期率']])
```

```python
def_agg['逾期天数'] = def_agg['还款日天数'] - def_agg['到期日天数']
def_agg.loc[def_agg['逾期天数'] < 0, '逾期天数'] = 0
def_agg.loc[(def_agg['到期日天数']> -30) & (def_agg['还款日天数'].isnull()), '逾期天数'] = 0
def_agg['逾期天数'].fillna(0, inplace=True)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

**数据合并与预处理**  
在机器学习的开始，通常是确保所有相关的数据都被整合到一个表格中。这里，我们将历史的借贷记录与逾期记录结合起来，提供一个全面的视角来分析每个借款人的信贷历史。合并后，某些字段可能会有空值，我们选择用一个特殊标志（例如-1）来填充这些空值，确保之后的分析不会受到干扰。

```python
# 1. 数据合并
# 将历史借贷记录和逾期记录并入总表
data_comb = pd.merge(data, def_data[['本期ID', '合计逾期天数']], on = '本期ID', how = 'left')
data_comb = pd.merge(data_comb, his_data[['本期ID', '历史借贷次数']], on = '本期ID', how = 'left')
data_comb.info()
```

```python
# 合计逾期天数和金额里有空值，为表示其特殊性，将其设为-1
data_comb['合计逾期天数'].fillna(-1, inplace = True)
data_comb.head()
```

**特征工程**   
定义类别变量、连续变量和目标变量，并进行类别变量的转换：  
为了让机器学习算法能够更好地处理这些类别变量数据（如性别、学历等），我们需要将它们转化为数值型数据。这里我们选择了一种常用的方法：one-hot编码。

```python
# 定义类别变量名称
categorical_features = [
    '性别', '学历', '婚姻状况', '职业类别', '年龄分箱', '是否有车', '是否有房'
]
# 定义连续变量名称
continous_features = [
    '年收入log', '家庭人数', '总贷款额度log', '征信查询次数', '工龄', '合计逾期天数', '历史借贷次数'
]
# 定义目标变量名称
target = "是否逾期"

# 将数据改成one-hot编码
X = pd.get_dummies(data_comb[continous_features + categorical_features], columns =categorical_features)
y = data_comb[target]
```

**数据集分割**  
为了评估模型的表现，我们需要将数据分为训练集和测试集。这里我们使用了train_test_split方法，并确保了训练集和测试集中的逾期用户比例与原数据接近。

```python
# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 1)
print('训练集里逾期用户占比: {:.1%}'.format(np.bincount(y_train)[1] / len(y_train)))
print('测试集里逾期用户占比: {:.1%}'.format(np.bincount(y_test)[1] / len(y_test)))
```

#### 2、逻辑回归 (Logistic Regression)

逻辑回归是金融风控领域中最常用的算法之一。尽管它是一个线性模型，但其背后的数学理论使其非常适合估计事件的概率，例如贷款违约的可能性。金融机构尤其喜欢使用逻辑回归，因为其模型的解释性强，能够直观地解释各个特征与目标事件（如逾期）之间的关系。这有助于构建风控策略并与监管机构进行沟通。

```python
# ========== 逻辑回归模型 ==========
# 1. 导入所需库并建立模型
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=10000)  # 使用了一个较大的迭代次数，确保模型收敛
logreg.fit(X_train, y_train)  # 模型拟合

# 2. 预测与模型评估
# 使用平均逾期率作为决策阈值进行预测
threshold = data[target].mean()

y_prob = logreg.predict_proba(X_test)  # 预测每个样本为各类的概率
y_pred = (y_prob[:,1] > threshold)  # 使用自定义阈值进行预测

# 输出模型的评估指标
print('逻辑回归模型的准确率为：', accuracy_score(y_test, y_pred))
print('逻辑回归模型f1分数为：', f1_score(y_test, y_pred))
print('逻辑回归模型召回值分数为：', recall_score(y_test, y_pred))
print('逻辑回归模型准确率分数为：', precision_score(y_test, y_pred))
print('逻辑回归模型AUC分数为：', roc_auc_score(y_test, y_pred))

# 3. 绘制混淆矩阵
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(  # 使用sklearn的工具绘制混淆矩阵
    confusion_matrix=cm, 
    display_labels=['正常', '逾期']
)
disp.plot()
```

#### 3、随机森林 (Random Forest)  
随机森林是一个集成学习方法，它结合了多个决策树的预测结果，通常能提供比单个决策树更精确的结果。在风控中，随机森林可以捕捉到复杂的非线性关系，并对不同的数据具有良好的稳定性。此外，随机森林能够评估每个特征的重要性，这对于确定哪些因素最能预测违约风险尤为有价值。

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

model = RandomForestClassifier(random_state=42) # 初始化随机森林模型

params = { # 通过网格搜索，寻找最优参数
    "n_estimators": [100, 200, 300], # 决策树的数量
    "max_depth": [3, 4, 5] # 决策树的最大深度
}

rf_classifier = GridSearchCV( # 初始化网格搜索
    estimator = model, # 模型
    param_grid = params, # 参数
    scoring = 'roc_auc', # 评价指标
    cv = 5 # 交叉验证的折数
)

rf_classifier.fit(X_train, y_train) # 模型训练
print(rf_classifier.best_params_) # 输出最优参数
```

```python
# testing result
threshold = data[target].mean() # 定义阈值为均值

y_prob = rf_classifier.predict_proba(X_test) # 预测概率
y_pred = (y_prob[:,1] > threshold) # 概率大于阈值的预测为1，否则为0

print('随机森林模型的准确率为：', accuracy_score(y_test, y_pred)) # 计算准确率
print('随机森林模型f1分数为：', f1_score(y_test, y_pred)) # 计算f1分数
print('随机森林模型召回值分数为：', recall_score(y_test, y_pred))  # 计算召回值分数
print('随机森林模型准确率分数为：', precision_score(y_test, y_pred)) # 计算准确率分数
print('随机森林模型AUC分数为：', roc_auc_score(y_test, y_pred)) # 计算AUC分数

cm = confusion_matrix(y_test, y_pred) # 计算混淆矩阵
disp = ConfusionMatrixDisplay( # 绘制混淆矩阵图
    confusion_matrix = cm, 
    display_labels = ['正常', '逾期']
)
disp.plot() # 绘制混淆矩阵图
```

#### 4、XGBoost
XGBoost 是一个优化的梯度提升机框架，近年来在各种数据科学比赛中取得了卓越的表现。在金融风控中，XGBoost 由于其出色的预测能力、处理大规模数据的速度以及能够自动处理缺失值等特点而受到赞誉。它可以处理非线性关系，并且像随机森林一样，可以为我们提供特征的重要性评分。不过，与逻辑回归相比，它的解释性可能会稍逊一筹。

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

model = xgb.XGBClassifier(random_state=42) # 初始化XGBoost模型

params = { # 通过网格搜索，寻找最优参数
    "n_estimators": [100, 200, 300], # 估计器个数
    "learning_rate": [0.1, 0.2, 0.3], # 学习率
    "max_depth": [3, 4, 5] # 树的最大深度
}

xgb_classifier = GridSearchCV( # 初始化网格搜索
    estimator = model, # 模型
    param_grid = params, # 参数
    scoring = 'roc_auc', # 评价指标
    cv = 5 # 交叉验证折数
)

xgb_classifier.fit(X_train, y_train) # 模型训练
print(xgb_classifier.best_params_) # 输出最优参数
```

```python
# testing result
threshold = data[target].mean() # 定义阈值为均值

y_prob = xgb_classifier.predict_proba(X_test) # 预测概率
y_pred = (y_prob[:,1] > threshold) # 概率大于阈值的预测为1，否则为0

print('XGBoost模型的准确率为：', accuracy_score(y_test, y_pred)) # 计算准确率
print('XGBoost模型f1分数为：', f1_score(y_test, y_pred)) # 计算f1分数
print('XGBoost模型召回值分数为：', recall_score(y_test, y_pred))  # 计算召回值分数
print('XGBoost模型准确率分数为：', precision_score(y_test, y_pred)) # 计算准确率分数
print('XGBoost模型AUC分数为：', roc_auc_score(y_test, y_pred)) # 计算AUC分数

cm = confusion_matrix(y_test, y_pred) # 计算混淆矩阵
disp = ConfusionMatrixDisplay( # 绘制混淆矩阵图
    confusion_matrix = cm, 
    display_labels = ['正常', '逾期']
)
disp.plot() # 绘制混淆矩阵图
```
