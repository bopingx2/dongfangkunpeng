```python
import requests
import pandas as pd

def get_stock_daily(code, start_date, end_date):
    url = f'https://datacenter.finedu.cc/api/stock/daily?code={code}&start_date={start_date}&end_date={end_date}'
    resp = requests.get(url=url)
    data = resp.json()
    return pd.DataFrame(data=data['data'])
```
